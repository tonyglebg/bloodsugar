```python
def get_info(root_dir: str) -> pd.DataFrame:
    df = pd.DataFrame(columns=['Name', 'block', 'group', 'PPG'])
    for subdir, dirs, files in tqdm(os.walk(rootdir)):
        for file in files:
            path = os.path.join(subdir, file)
            if path.endswith('.txt') and ('Vegetatics' in path):
                group = ''
                name = ''
                block = ''
                PPG = []
                if "Control" in path: 
                    group = 'Control'
                elif 'Wrong_feedback' in path:
                    group = 'Wrong_feedback'
                name = file
                block = file.split("_")[2][:-4]
                print(file)
                try:
                    regex = re.compile(r"-?[0-9][.]?[0-9]*")
                    i = 0
                    with open(path) as f:
                        for line in f:
                            if i < 47:
                                i+=1
                            else:
                                value = regex.findall(line)
                                if len(value) != 0:
                                    PPG.append(float(value[3]))
                                else:
                                    continue
                except:
                    continue
                row = {'Name': name, 'block': block, 'group': group, 'PPG': PPG}
                df = df.append(row, ignore_index=True)
    return df
```

```python
import numpy as np
import scipy.signal as signal
fs = 250  
lowcut = 0.5
highcut = 4

b, a = signal.butter(2, [lowcut, highcut], btype='bandpass', fs=fs)
filtered_ppg = signal.filtfilt(b, a, qwe)


ppg_derivative = np.gradient(filtered_ppg)
zero_crossings = np.where(np.diff(np.sign(ppg_derivative)))[0]

notches = []
for index in zero_crossings:
    if filtered_ppg[index] > filtered_ppg[index + 1]:
        notch = index
    else:
        notch = index + 1
    notches.append(notch)
print(notches)
```
